```python
import pandas as pd
df = pd.read_csv("Lemonade.csv")
df.head(10)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date',inplace=True)
```

```python
print(df)
```

```python
Day = df['Day'].values
Temperature = df['Temperature'].values
Rainfall = df['Rainfall'].values
Flyers = df['Flyers'].values
Price = df['Price'].values
Sales = df['Sales'].values
```

```python
print(df['Day'])
print(df['Temperature'])
print(df['Rainfall'])
print(df['Flyers'])
print(df['Price'])
print(df['Sales'])
```

```python
import numpy as np
mean = np.mean(Temperature)
median = np.median(Temperature)
print(mean)
print(median)
```

```python
mean = np.mean(Rainfall)
median = np.median(Rainfall)
print(mean)
print(median)
```

```python
mean = np.mean(Flyers)
median = np.median(Flyers)
print(mean)
print(median)
```

```python
mean = np.mean(Price)
median = np.median(Price)
print(mean)
print(median)
```

```python
mean = np.mean(Sales)
median = np.median(Sales)
print(mean)
print(median)
```

```python
revenue = np.multiply(Sales,Price)
revenue[0]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(Sales,revenue)
```

```python
plt.xlabel("Sales")
plt.ylabel("revenue")
plt.show()
```
