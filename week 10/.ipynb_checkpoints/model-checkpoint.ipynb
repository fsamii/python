```python
import pandas as pd
```

```python
df = pd.read_csv('googleplaystore.csv')
# df
df1 = df.dropna()
df1
```

```python
#   Question-1
```

```python
# counting unique items
item_counts = df1["Category"].value_counts()
print(item_counts)
```

```python
item_counts.count()
```

```python
rating = df1.groupby('Category').mean()
print(rating)
```
